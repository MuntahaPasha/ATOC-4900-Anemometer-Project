**IMPORTS NEEDED**

Import these Python Libraries to use within this Notebook.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
from windrose import WindroseAxes
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import csv
```

**WIND ROSE FOR WXT (2018 AND 2019)**

First lets read in the dataframes and output the information.

Let's start with 2018.

```python
dfWxT2018 = pd.read_csv('WXT2018', encoding='utf-8')
dfWxT2018.info()
```

Now let's look at our 2019 one.

```python
dfWxT2019 = pd.read_csv('WXT2019', encoding='utf-8')
dfWxT2019.info()
```

And our 2020 data as well.

```python
dfWxT2020 = pd.read_csv('WXT2020', encoding='utf-8')
dfWxT2020.info()
```

Now we can plot up a windrose for 2018.

```python
windspdWxT2018 = dfWxT2018['wspd'].tolist()
winddirWxT2018 = dfWxT2018['wdir'].tolist()
ax = WindroseAxes.from_ax()
bins_range = np.arange(0, 20, 4.9)
ax.bar(winddirWxT2018, windspdWxT2018, normed=True, opening=0.8, edgecolor='white', bins=bins_range)
ax.set_legend()
ax.set_title('Wind Rose for WXT 2018 Data')
```

And let's do the same thing for 2019.

```python
windspdWxT2019 = dfWxT2019['wspd'].tolist()
winddirWxT2019 = dfWxT2019['wdir'].tolist()
ax = WindroseAxes.from_ax()
bins_range = np.arange(0, 20, 4.9)
ax.bar(winddirWxT2019, windspdWxT2019, normed=True, opening=0.8, edgecolor='white', bins=bins_range)
ax.set_legend()
ax.set_title('Wind Rose for WXT 2019 Data')
```

Also for 2020.

```python
windspdWxT2020 = dfWxT2020['wspd'].tolist()
winddirWxT2020 = dfWxT2020['wdir'].tolist()
ax = WindroseAxes.from_ax()
bins_range = np.arange(0, 20, 4.9)
ax.bar(winddirWxT2020, windspdWxT2020, normed=True, opening=0.8, edgecolor='white', bins=bins_range)
ax.set_legend()
ax.set_title('Wind Rose for WXT 2020 Data')
```

**WINDROSE CONCATENATION FOR 2018, 2019, AND 2020**

As requested, I concatenated the WxT data for all the years we have on file.

```python
frames = [dfWxT2018, dfWxT2019, dfWxT2020]
#concatenate dataframes
dfWXT = pd.concat(frames, sort=False)
dfWXT.info()
```

**ALL WXT DATA WINDROSE**

This is a Complete Windrose of all WXT data that we have on file.

```python
#windspdWxT = dfWXT['wspd'].tolist()
#winddirWxT = dfWXT['wdir'].tolist()
ax = WindroseAxes.from_ax()
bins_range = np.arange(0, 45, 5)
ax.bar(dfWXT['wdir'], dfWXT['wspd'], normed=True, opening=0.8, edgecolor='white', bins=bins_range)
ax.set_legend()
ax.set_title('Wind Rose for WXT Data')
```
