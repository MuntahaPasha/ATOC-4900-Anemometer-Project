**READING IN THE 2018 SONIC DATA**

I used a Pandas Dataframe here and printed the information contained within the dataframe after reading it in.

```python
import pandas as pd
df = pd.read_csv('2018Sonic', encoding='utf-8')
df.info()
```

**SPLIT-PLOTTING FOR 2018 SONIC DATA (PART 1)**

I noticed there were about 3 million data points, so to make processing easier, I cut it down into slices and plotted the first million data points, and then the rest. I am plotting here the first 1 million data points.

```python
ax = df.iloc[0:10000000].plot(kind = 'line', x = 'datetime', y = 'Spd', color='blue', figsize = (20, 10))
ax.set_xlabel("Time (Half-Year)")
ax.set_ylabel("Wind Speed (m/s)")
```

**SPLIT-PLOTTING FOR 2018 SONIC DATA (PART 2)**

I am plotting here the rest of the data points. (Didn't do an even split here)

```python
ax2 = df.iloc[1000000:29410107].plot(kind = 'line', x = 'datetime', y = 'Spd', color='navy', figsize = (20, 10))
ax2.set_xlabel("Time (Half-Year)")
ax2.set_ylabel("Wind Speed (m/s)")
```

**PLOTTING ALL OF SONIC 2018 DATA**

I decided not to use a split here, and just plot all of the 2018 data in one chart. It takes about a minute or so to run.

```python
ax = df.plot(kind = 'line', x = 'datetime', y = 'Spd', color='magenta', figsize = (20, 10))
ax.set_xlabel("Time Elapsed in 2018")
ax.set_ylabel("Wind Speed (m/s)")
```

**READING IN THE 2019 SONIC DATA**

I used a Pandas Dataframe here and printed the information contained within the dataframe after reading it in.

```python
df2 = pd.read_csv('2019Sonic', encoding='utf-8')
df2.info()
```

**SPLIT-PLOTTING FOR 2019 SONIC DATA (PART 1)**

I noticed that in this file there were more than 3 million data points, so again I cut it down into slices and plotted the first million and a half data points, and then the rest. Here I did an even split.

```python
ax3 = df2.iloc[0:15715651].plot(kind = 'line', x = 'datetime', y = 'Spd', color='lime', figsize = (20, 10))
ax3.set_xlabel("Time (Half-Year)")
ax3.set_ylabel("Wind Speed (m/s)")
```

**SPLIT-PLOTTING FOR 2019 SONIC DATA (PART 2)**

I am plotting here the rest of the data points.

```python
ax4 = df2.iloc[15715651:31431303].plot(kind = 'line', x = 'datetime', y = 'Spd', color='green', figsize = (20, 10))
ax4.set_xlabel("Time (Half-Year)")
ax4.set_ylabel("Wind Speed (m/s)")
```

**PLOTTING ALL OF SONIC 2019 DATA**

I decided not to use a split here, and just plot all of the 2019 data in one chart. It takes about a minute or so to run.

```python
ax = df2.plot(kind = 'line', x = 'datetime', y = 'Spd', color='brown', figsize = (20, 10))
ax.set_xlabel("Time Elapsed in 2019")
ax.set_ylabel("Wind Speed (m/s)")
```

**PLOTTING SONIC 2018 DATA WITH 2019 DATA**

The red line chart is $\textbf{2018}$ data, and the blue is $\textbf{2019}$ data.

```python
#2018 is red
ax = df.plot(kind = 'line', x = 'datetime', y = 'Spd', color='red', figsize = (20, 10))
#2019 is blue
df2.plot(kind = 'line', x = 'datetime', y = 'Spd', figsize = (20, 10), ax = ax)
```

**CONCATENATING DATA**

Now, I decided to combine the data from 2018 with the data from 2019 into one continuous graph.

```python
frames = [df, df2]
#concatenate dataframes
df20182019 = pd.concat(frames, sort=False)
```

I'm just printing out some information about the new dataframe to see if it merged the points correctly.

```python
df20182019.info()
```

**PLOTTING CONCATENATED DATA**

I finally plotted my graph, and as you can see it's one continuous chart from 2018 to 2019.

```python
ax = df20182019.plot(kind = 'line', x = 'datetime', y = 'Spd', color='gold', figsize = (20, 10))
```
